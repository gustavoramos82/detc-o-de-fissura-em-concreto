```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
epocas = [1,2,3,4,5]

loss_vgg16 = [0.2535,0.1222, 0.0991,0.0466, 0.0453]
loss_resnet50 = [0.1789,0.0319,0.0218,0.0243, 0.0958]
loss_densenet121 = [0.1496,0.0673,0.0224,0.0156,0.0179]
loss_mobiletnet = [0.2156,0.0159, 0.0094,0.0101, 0.1615]
```

```python
plt.figure(figsize=(12,6))
plt.title('Função perdas das redes ao longo do treinamento',fontsize=14)

plt.plot(epocas,loss_vgg16,color='red',label='VGG16')
plt.plot(epocas,loss_resnet50,color='blue',label='ResNet50')
plt.plot(epocas,loss_densenet121,color='green',label='DenseNet121')
plt.plot(epocas,loss_mobiletnet,color='purple',label='MobileNet')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()
```

```python
acc_vgg16 = [0.9117,0.9577,0.9660,0.9857,0.9860]
acc_resnet50 = [ 0.9522,0.9900,0.9940,0.9921,0.9644]
acc_densenet121 = [0.9596,0.9775,0.9938,0.9953,0.9942]
acc_mobilenet = [0.9623, 0.9954,0.9972,0.9969,0.9712]
```

```python
plt.figure(figsize=(12,6))
plt.title('Acurácia das redes ao longo do treinamento',fontsize=14)

plt.plot(epocas,acc_vgg16,color='red',label='VGG16')
plt.plot(epocas,acc_resnet50,color='blue',label='ResNet50')
plt.plot(epocas,acc_densenet121,color='green',label='DenseNet121')
plt.plot(epocas,acc_mobilenet,color='purple',label='MobileNet')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.legend()
```

```python
tempo_treinamento = [3.03,4.5,6.91,1.76]
redes = ['VGG16','ResNet50','DenseNet121','Mobilenet']

plt.figure(figsize=(10,6))
plt.title('Tempo de treinamento das redes',fontsize=14)
plt.bar(redes,tempo_treinamento)
plt.ylabel('Tempo (min)')
```
